```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("G:/Education/Machine Learning/ML_CSV_FILES/Market_Basket_Optimisation.csv", header = None)
df.head()
```

```python
transactions = [] # SQUARE BRACES CREATE EMPTY LIST#
for i in range(0,7501):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
```

```python
# IN APYORI WE NEED TO CREATE RULES IN ORDER TO CHECK & CLUB FOR THE RECOMMENDATIONS OF THE PRODUCT WISHLISTED OR BOUGHT #
# MIN_LENGTH REFERS TO THE ITEM RECOMMENDATIONS NEED TO BE DISPLAYED #
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
# THE CREATED RULES WILL CONTAIN THE RECOMMENDATIONS LISTED IN THE "for loop" THE FOR LOOP WILL CREATE THE RULES AND RECOMMENDATIONS #
results = list(rules)
results
```
