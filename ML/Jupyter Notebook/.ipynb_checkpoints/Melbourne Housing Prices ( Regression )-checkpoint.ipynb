```python
df.drop(['Postcode', 'Longtitude', 'Lattitude', 'Regionname', 'BuildingArea', 'Propertycount', 'Landsize', 'YearBuilt', 'CouncilArea'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
len(df["Date"].unique())/4 # 12 represents the data related to an year. #
```

```python
df[df["Type"] == "h"].sort_values("Date", ascending = False).groupby("Date").mean()
```

```python
pd.set_eng_float_format(accuracy=1, use_eng_prefix=True)
df[(df["Type"] == "h") & 
   (df["Distance"]<14) &
   (df["Distance"]>13.5)].sort_values("Date", ascending = False).dropna().groupby(["Suburb", "SellerG"]).mean()
```

```python
import math
Bedroom_Mean = math.floor(df['Bedroom2'].mean())
Bathroom_Mean = math.floor(df['Bathroom'].mean())
Car_Mean = math.floor(df['Car'].mean())
```

```python
df.Bedroom2 = df['Bedroom2'].fillna(Bedroom_Mean)
df.Bathroom = df['Bathroom'].fillna(Bathroom_Mean)
df.Car = df['Car'].fillna(Car_Mean)
```

```python
df.dtypes
```

```python
df.drop(['Suburb', 'Address', 'Method', 'SellerG', 'Date'], axis = 1, inplace = True)
```

```python
df['Type'] = pd.get_dummies(df['Type'])
```

```python
df.dtypes
```

```python
x = df.drop(['Price'], axis = 1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Regression Analysis

```python
from sklearn.metrics import accuracy_score, classification_report, r2_score
from sklearn import metrics
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
prediction = regression.predict(x_test)
```

```python
score = round((regression.score(x_test, y_test)*100),2)
print("\nModel Score:",score,"%")
```

```python
r2_score(y_test,prediction)
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=300, random_state=0)
regressor.fit(x,y)
```

```python
prediction =regressor.predict(x_test)
```

```python
score = round((regressor.score(x_test, y_test)*100),2)
print("\nModel Score:", score,"%")
```

```python
r2_score(y_test,prediction)
```

## Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel='rbf')
regressor.fit(x,y)
```

```python
prediction = regressor.predict(x_test)
```

```python
score = round((regressor.score(x_test, y_test)*100),2)
print("\nModel Score:", score,"%")
```

```python
r2_score(y_test,prediction)
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(x,y)
```

```python
prediction = regressor.predict(x_test)
```

```python
score = round((regressor.score(x_test, y_test)*100),2)
print("\nModel Score:",score,"%")
```

```python
r2_score(y_test,prediction)
```

### Major Deciding Factor of House price

```python
classifier = RandomForestRegressor()
classifier.fit(x,y)
```

```python
feature_names = x.columns
importance_frame = pd.DataFrame()
importance_frame['Features'] = x.columns
importance_frame['Importance'] = classifier.feature_importances_
importance_frame = importance_frame.sort_values(by=['Importance'], ascending=True)
```

```python
plt.barh([0,1,3,4,5,6], importance_frame['Importance'], align='center', alpha=0.5)
plt.yticks([0,1,3,4,5,6], importance_frame['Features'])
plt.xlabel('Importance')
plt.title('Deciding Factor')
plt.show()
```

## The major deciding factor in the price mappings of melbourne is Distance
