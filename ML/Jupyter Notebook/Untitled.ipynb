```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
df=pd.read_csv('C:/Users/varun/Downloads/Data.csv')
df.head()
```

```python
df.set_index('consumed_at')
```

```python
df['consumed_at']=pd.to_datetime(df['consumed_at'])
```

```python
df_final=pd.DataFrame(columns=['Datetime FROM','Datetime TO','Count','ActivePower1 SUM','ActivePower2 SUM','ActivePower3 SUM','Watts','Units'])

for i in range(0,24):
    if i == 0:
        for index, row in df.iterrows():
            if row['consumed_at'].hour == 0 and row['consumed_at'].minute == 0:
                startdate = row['consumed_at']
                enddate = startdate + timedelta(hours = 1)
                df_hr = df[df.consumed_at.between(startdate, enddate)]
                print(df_hr)
                print("-------------------------")

                count = df_hr['active_power_1'].count()
                watt_1 = df_hr['active_power_1'].sum()
                watt_h_1 = watt_1/count
                kwh_1 = watt_h_1/1000

                watt_2=df_hr['active_power_2'].sum()
                watt_h_2=watt_2/count
                kwh_2=watt_h_2/1000

                watt_3=df_hr['active_power_3'].sum()
                watt_h_3=watt_3/count
                kwh_3=watt_h_3/1000

                vallist = [[startdate, enddate, count, watt_1, watt_2, watt_3, (watt_h_1+watt_h_2+watt_h_3), (kwh_1+kwh_2+kwh_3)]]
                df_final = df_final.append(pd.DataFrame(vallist, columns=['Datetime FROM','Datetime TO','Count','ActivePower1 SUM','ActivePower2 SUM','ActivePower3 SUM','Watts','Units']),ignore_index=True)
                df_final = df_final.drop_duplicates()
                print("DF FINAL")
                print(df_final)
        
    else:
        startdate = enddate
        enddate = startdate + timedelta(hours = 1)
        print(startdate, "_", enddate)
        df_hr = df[df.consumed_at.between(startdate, enddate)]
        print(df_hr)
        print("-------------------------")
        
        count = df_hr['active_power_1'].count()
        watt_1 = df_hr['active_power_1'].sum()
        watt_h_1 = watt_1/count
        kwh_1 = watt_h_1/1000
        
        watt_2=df_hr['active_power_2'].sum()
        watt_h_2=watt_2/count
        kwh_2=watt_h_2/1000

        watt_3=df_hr['active_power_3'].sum()
        watt_h_3=watt_3/count
        kwh_3=watt_h_3/1000
        
        vallist = [[startdate, enddate, count, watt_1, watt_2, watt_3, (watt_h_1+watt_h_2+watt_h_3), (kwh_1+kwh_2+kwh_3)]]
        df_final = df_final.append(pd.DataFrame(vallist, columns=['Datetime FROM','Datetime TO','Count','ActivePower1 SUM','ActivePower2 SUM','ActivePower3 SUM','Watts','Units']),ignore_index=True)
        print("DF FINAL")
        print(df_final)
```
