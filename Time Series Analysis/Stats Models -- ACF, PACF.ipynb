# Example of acf, pacf, pacf_yw, pacf_ols

```python
df = pd.DataFrame({'a':[10,20,2,96,4,604,27]})
```

```python
df
```

```python
# Auto Correlation Function (acf)
acf(df['a'])
```

```python
# Partial Auto Correlation Function_Youl Walker Equation (pacf_yw)
# method = mle; stands for 'MAXIMUM LIKELIHOOD ESTIMATION' to use the unbiased auto correlation function (acf) else we can type 'unbiased'
pacf_yw(df['a'], nlags = 6, method = 'mle') # nlags = n-1 i.e; we have total of 7 rows and 7-1 = 6
```

```python
# Partial Auto Correlation Function_Ordinary Least Square Estimates
pacf_ols(df['a'], nlags = 6)
```
