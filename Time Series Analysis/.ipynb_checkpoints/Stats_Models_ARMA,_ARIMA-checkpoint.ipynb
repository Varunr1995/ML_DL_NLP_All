## Checking whether the data is stationary or not, by using Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df1['Births'])
```

```python
auto_arima(df1['Births'],seasonal = False).summary()
```

```python
# Predicting One month data
start = len(train)
end = len(train)+len(test) - 1
```

```python
predictions = results.predict(start, end).rename('ARMA(2,2) Predictions')
predictions
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df2['Inventories'], model = 'add')
result.plot();
```

* The reason to provide seasonal = False eventhough we see seasonality with naked eye, is to confirmly verify by doing the auto_arima and dickey-fuller test and confirm.

```python
auto_arima(df2['Inventories'], seasonal = False).summary()
```

* Even though we provided seasonal=False; the summary provided that the data do have some seasonality and provided it in ARIMA order.

```python
from statsmodels.tsa.statespace.tools import diff
```

```python
train = df2.iloc[:252].astype('float32')
test = df2.iloc[252:].astype('float32')
```

```python
model = ARIMA(train['Inventories'], order = (0, 1, 0)) # The order is given by the best order provided by auto_arima.
results = model.fit()
results.summary()
```

```python
start = len(train)
end = len(train) + len(test) - 1
```

```python
print('start : ', start)
print('end : ', end)
```

```python
predictions = results.predict(start = start, end = end, typ = 'levels').rename('ARIMA (0,1,0) Predictions')  # The type(typ) will be either linear or levels.
# linear : Linear prediction in terms of the differenced endogenous variables.
# levels : predict the levels of the original endogenous variable.
```
