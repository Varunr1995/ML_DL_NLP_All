## Sentiment Analysis

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def print_sentiment_scores(sentence):
    snt = analyzer.polarity_scores(sentence)
    print("{:-<40}{}".format(sentence, str(snt)))
```

```python
print_sentiment_scores("This raspberry cake is good.")
print_sentiment_scores("This raspberry cake is GOOD!")
print_sentiment_scores("This raspberry cake is GOOD!!")
print_sentiment_scores("This raspberry cake is VERY GOOD!!")
```

#### Printing Sentiment Scores Individually.

```python
def negative_score(text):
    negative_value = analyzer.polarity_scores(text)['neg']
    return negative_value

def positive_score(text):
    positive_value = analyzer.polarity_scores(text)['pos']
    return positive_value

def neutral_score(text):
    neutral_value = analyzer.polarity_scores(text)['neu']
    return neutral_value

def compound_score(text):
    compound_value = analyzer.polarity_scores(text)['compound']
    return compound_value
```

```python
print("Negative Score: ",negative_score("The food tasted bad and the was also bad!!!"))
print("Neutral Score: ",neutral_score("The food tasted bad and the was also bad!!!"))
print("Positive Score: ",positive_score("The food tasted bad and the was also bad!!!"))
print("Compound Score: ",compound_score("The food tasted bad and the was also bad!!!"))
```

### Calculating Sentiment Scores of English only!

```python
%%time

df_eng['sentiment_neg'] = df_eng['comments'].apply(negative_score)
df_eng['sentiment_neu'] = df_eng['comments'].apply(neutral_score)
df_eng['sentiment_pos'] = df_eng['comments'].apply(positive_score)
df_eng['sentiment_com'] = df_eng['comments'].apply(compound_score)
```

## Topic Modelling

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
```

```python
def clean(doc):
    stop_free =" ".join([word for word in doc.lower().split() if word not in stop])
    punct_free ="".join(token for token in stop_free if token not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punct_free.split())
    return normalized

doc_clean = [clean(comment).split() for comment in pos_comments]
```

#### First we need to create a gensim dictionary from the normalized_data and converting in to bag-of-words corpus for further use.

```python
dictionary = corpora.Dictionary(doc_clean)
corpus = [dictionary.doc2bow(text) for text in doc_clean]
```

```python
pickle.dump(corpus, open('NLP_SentimentData_Corpus.pkl', 'wb'))
dictionary.save('NLP_SentimentData_Dictionary.gensim')
```
