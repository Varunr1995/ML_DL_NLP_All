## Finding Frequenct Distribution

```python
from nltk.probability import FreqDist
```

```python
freq_dist = FreqDist(words)
freq_dist.most_common(10)
```

### The most used words are punctuation marks and stopwords. We will have to remove such words to analyze the actual text.

## Removing Punctuation Marks

### isalpha is a built in library of nlp in order to remove punctuation marks and convert whole text in to lowercase.

```python
words_without_punct = []
```

```python
for w in words:
    if w.isalpha():
        words_without_punct.append(w.lower())
```

```python
print(words_without_punct)
```

```python
freq_dist = FreqDist(words_without_punct)
freq_dist.most_common(10)
```

## List of Stop Words

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
print(stop_words)
```

## Removing Stopwords

```python
cleaned_words = []
```

```python
for w in words_without_punct:
    if w in stop_words:
        cleaned_words.append(w)
```

```python
print(cleaned_words)
print("\n")
print(len(cleaned_words))
```

## Final Frequency Distribution

```python
fdist = FreqDist(cleaned_words)
fdist.most_common(10)
```

## Named Entity Recognisation

```python
sentence = "Mr. Smith works at The New Canyon Company located at New York city"
```

```python
tokenized_sent = word_tokenize(sentence)
```

```python
for w in tokenized_sent:
    tagged_words = nltk.pos_tag(tokenized_sent)
    
ner = nltk.ne_chunk(tagged_words, binary = False)
print(ner)
ner.draw()
```

## Word Net

### We can use Wordnet to find meanings of words, synonyms, antonyms, and many other words.

```python
from nltk.corpus import wordnet
```

```python
for words in wordnet.synsets("Fun"):
    print(words)
```

```python
for words in wordnet.synsets("Fun"):
    for lemma in words.lemmas():
        print(lemma)
    print('\n')
```

###  All details for a word.

```python
word = wordnet.synsets("Play")[0]
print(word.name())
print('\n')
print(word.definition())
print('\n')
print(word.examples())


```

```python
for words in wordnet.synsets("Fun"):
    print(words.name())
    print(words.definition())
    print(words.examples())
    
    for lemma in words.lemmas():
        print(lemma)
    print('\n')
```

## Synonyms

```python
synonyms = []

for words in wordnet.synsets('Fun'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        
print(synonyms)
```

## Antonyms

```python
antonyms = []

for words in wordnet.synsets('dead'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
        
print(antonyms)
```

## Synonyms & Antonyms

```python
synonyms = []
antonyms = []

for words in wordnet.synsets("Happy"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
            
print("Synonyms:", synonyms)
print('\n')
print('Antonyms:', antonyms)
```

## Finding Similarity B/W Words

```python
word1 = wordnet.synsets('ship', 'n')[0]
word2 = wordnet.synsets('cruise', 'n')[0]
print(word1.wup_similarity(word2))
```

## Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences = ["Jim and Cary travelled to hospital in car",
             "Doctor arrived on time to treat injured Symon",
             "Symon is mutual friend of Jim and Cary"]
```

```python
cv = CountVectorizer()
```

```python
B_O_W = cv.fit_transform(sentences).toarray()
print(cv.vocabulary_)
print('\n')
print(cv.get_feature_names())
print('\n')
print(B_O_W)
```

## TF-IDF basic working syntax

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences = ['This is the first document',
             'This document is the second document']
```

```python
vectorizer = TfidfVectorizer(norm = None)
```

```python
x = vectorizer.fit_transform(sentences).toarray()
```

```python
print(vectorizer.vocabulary_)
print('\n')
print(vectorizer.get_feature_names())
print('\n')
print(x)
```
